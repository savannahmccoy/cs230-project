```python
from typing import Dict, Text

import numpy as np
import csv
import pandas as pd
import tensorflow as tf
import time
import datetime
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow_datasets as tfds
import tensorflow_recommenders as tfrs
```

```python
# read data from csv files
reviews_df = pd.read_csv('reviews_categorical.csv')
reviews_df.columns = ["userID", "productID", "vote", "rating", "verified"]
reviews_df = reviews_df.sample(frac=0.25, replace=False, random_state=23)
product_df = pd.read_csv('meta_categorical.csv')
product_df = product_df.sample(frac=0.25, replace=False, random_state=23)
```

```python
# convert dataframe to tensor dataset format
reviews = tf.data.Dataset.from_tensor_slices(dict(reviews_df))
products = tf.data.Dataset.from_tensor_slices(dict(product_df))
```

```python
# confirm conversion successful
for row in reviews.take(3):
  print(row)
for row in products.take(3):
  print(row)
```

```python
# select the features to map
reviews = reviews.map(lambda x: {
    "productID": x["productID"],
    "userID": x["userID"],
    "rating": x["rating"],
    "vote": x["vote"],
    "verified": x["verified"],
})
products = products.map(lambda x: x["productID"])
```

```python
# create vocabularies for embeddings
start1 = time.time()
start = time.time()
user_ids_vocabulary = tf.keras.layers.StringLookup(mask_token=None)
user_ids_vocabulary.adapt(reviews.map(lambda x: x["userID"]))
end = time.time()
print("USER ID VOCAB:", str(datetime.timedelta(seconds=end-start)))
start = time.time()
product_ids_vocabulary = tf.keras.layers.StringLookup(mask_token=None)
product_ids_vocabulary.adapt(reviews.map(lambda x: x["productID"]))
end = time.time()
print("PRODUCT ID VOCAB:", str(datetime.timedelta(seconds=end-start)))
```

```python
# define a recommender model
class ProductRecommendationModel(tfrs.Model):

  def __init__(self, 
               rating_weight: float, 
               retrieval_weight: float,
               user_model: tf.keras.Model, 
               product_model: tf.keras.Model,
               task: tfrs.tasks.Retrieval):
    super().__init__()

    # user and product representations
    self.user_model = user_model
    self.product_model = product_model


    # model using user and product embeddings to predict ratings
    self.rating_model = tf.keras.Sequential([
        tf.keras.layers.Dense(256, activation="relu"),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dense(1),
    ])

    # loss weights
    self.rating_weight = rating_weight
    self.retrieval_weight = retrieval_weight


    # retrieval tasks: factorization loss and RMSE
    self.rating_task: tf.keras.layers.Layer = tfrs.tasks.Ranking(
        loss=tf.keras.losses.MeanSquaredError(),
        metrics=[tf.keras.metrics.RootMeanSquaredError()],
    )
    self.retrieval_task: tf.keras.layers.Layer = tfrs.tasks.Retrieval(
        metrics=tfrs.metrics.FactorizedTopK(
            candidates=products.batch(128).map(self.product_model)
        )
    )

  
  def call(self, features: Dict[Text, tf.Tensor]) -> tf.Tensor:
    user_embeddings = self.user_model(features["productID"])
    product_embeddings = self.product_model(features["productID"])
    return (user_embeddings, product_embeddings, self.rating_model(
        tf.concat([user_embeddings, product_embeddings], axis=1)
        )
    )


  def compute_loss(self, features: Dict[Text, tf.Tensor], training=False) -> tf.Tensor:
    # define how the loss is computed
    ratings = features.pop("rating")
    user_embeddings, product_embeddings, rating_predictions = self(features)

    # compute loss for each task
    rating_loss = self.rating_task(labels=ratings, predictions=rating_predictions)
    retrieval_loss = self.retrieval_task(user_embeddings, product_embeddings)

    # combine them using the loss weights
    return (self.rating_weight * rating_loss + self.retrieval_weight * retrieval_loss)

```

```python
# define user and product models
user_model = tf.keras.Sequential([
    user_ids_vocabulary,
    tf.keras.layers.Embedding(user_ids_vocabulary.vocabulary_size(), 64)
])
product_model = tf.keras.Sequential([
    product_ids_vocabulary,
    tf.keras.layers.Embedding(product_ids_vocabulary.vocabulary_size(), 64)
])

# define objectives
task = tfrs.tasks.Retrieval(metrics=tfrs.metrics.FactorizedTopK(
    products.batch(128).map(product_model)
  )
)
```

```python
# create retrieval model
model = ProductRecommendationModel(
                                  user_model=user_model, 
                                   product_model=product_model, 
                                   task=task, 
                                   rating_weight=0.9, 
                                   retrieval_weight=0.1)
model.compile(optimizer=tf.keras.optimizers.Adagrad(0.7))

# cache datasets to use for eval
cached_train = reviews.shuffle(100_000).batch(8192).cache()
cached_test = reviews.batch(4096).cache()

# train model
start = time.time()
history = model.fit(cached_train, epochs=20)
end = time.time()
print("TIME TO TRAIN MODEL:", str(datetime.timedelta(seconds=end-start)))

# print result metrics
start = time.time()
model.evaluate(cached_test, return_dict=True)
end = time.time()
print("TIME TO EVALUATE MODEL:", str(datetime.timedelta(seconds=end-start)))
end1 = time.time()
print("\n\n\n ----------------------------------------------------- \n\nTOTAL TIME TO TRAIN MODEL:", str(datetime.timedelta(seconds=end1-start1)))
```

```python
# plot train accuracy over epochs
epochs = [i for i in range(30)]
plt.plot(epochs, history.history["factorized_top_k/top_100_categorical_accuracy"], label="accuracy")
plt.title("Accuracy vs epoch")
plt.xlabel("epoch")
plt.ylabel("Top-100 accuracy");
plt.legend()
```
