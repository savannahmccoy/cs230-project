```python
# META
# category (expand) (dummy) 
# brand (dummy) 
# price (ordinal) 
# productID (None)
```

```python
meta_df = pd.read_csv('meta_filtered.csv')
```

```python
meta_df.head(2)
```

```python
meta_df.dtypes
```

```python
meta_df['price'] = meta_df['price'].astype(str)
```

```python
list(meta_df['price'])
```

```python
# get rid 
def filter_prices(x):
    if x.startswith('$'):
        return x
    else:
        return ''
    
meta_df["clean_price"] = meta_df["price"].apply(lambda x: filter_prices(x))
```

```python
meta_df.head(5)
```

```python
le = preprocessing.LabelEncoder()
le.fit(meta_df["clean_price"])
new_price = le.transform(meta_df["clean_price"])
meta_df["cat_price"] = new_price
```

```python
meta_df.head(10)
```

```python
meta_df['cat'] = meta_df['category'].str.strip('[]').str.split(',')
```

```python
f = list(meta_df['cat'])
```

```python
import itertools
lst = [x for l in f for x in l]
print(len(lst))
print(len(set(lst)))
```

```python
def get_cat2(x):
    if len(x) >= 2:
        return x[1]
    else:
        return ''
    
def get_cat3(x):
    if len(x) >= 3:
        return x[2]
    else:
        return ''
    
meta_df["cat2"] = meta_df["cat"].apply(lambda x: get_cat2(x))
meta_df["cat3"] = meta_df["cat"].apply(lambda x: get_cat3(x))
```

```python
meta_df.head(5)
```

```python
meta_df = pd.get_dummies(meta_df, columns=['cat2'])
```

```python
# meta_df = pd.get_dummies(meta_df, columns=['cat3'])
```

```python
meta_df.head(5)
```

```python
print(meta_df.columns)
```

```python
del meta_df["cat"]
del meta_df["category"]
del meta_df["price"]
del meta_df["clean_price"]
```

```python
meta_df.head(5)
```

```python
meta_df = pd.get_dummies(meta_df, columns=['brand'])
```

```python
meta_df.head(5)
```

```python
meta_df.to_csv("meta_categorical.csv", index=False)
```

```python
def bin_rating(x):
    if x >= 4:
        return 1
    else:
        return 0

reviews_df["bin_rating"] = reviews_df["rating"].apply(lambda x: bin_rating(x))
```

```python
def bin_verified(x):
    if x:
        return 1
    else:
        return 0

reviews_df["bin_verified"] = reviews_df["verified"].apply(lambda x: bin_verified(x))
```
