```python
import csv
import pandas as pd
import json
```

```python
df = pd.read_csv("Video_Games_5.json", sep="\n", header=None)
```

```python
# df
```

```python
d = {}
def split_data(row):
    dt = json.loads(row[0])
    d[row.name] = dt
    return

df.apply(lambda row: split_data(row), axis=1)
# print(d)
```

```python
d[14000]
```

```python
with open("sample.json", "w") as outfile:
    json.dump(d, outfile)
```

```python
df2 = pd.DataFrame.from_dict(d)
```
